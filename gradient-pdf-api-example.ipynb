# PDFMiner Function

```python
file_loc = "sample_pdf.pdf"
```

```python
def extract_text_from_page(pdf_path, page_numbers):
    output_string = StringIO()
    with open(pdf_path, 'rb') as file:
        for page in PDFPage.get_pages(file, page_numbers):
            text = extract_text(file, laparams=LAParams(), page_numbers=[page.pageid - 1])
            output_string.write(text)
    return output_string.getvalue()
```

# PDF API Block with RAGs

```python
filepath = "attention.pdf"
```

```python
result = gradient.extract_pdf(
    filepath=filepath
)
```

```python
type(result["pages"])
```

```python
len(result["pages"])
```

```python
for page in range(len(result["pages"])):
    with open("Output/Page{}.txt".format(page), "w") as text_file:
        text_file.write(result["pages"][page]['text'])
```

```python
llm = GradientBaseModelLLM(
    base_model_slug="llama2-7b-chat",
    max_tokens=400,
)
```

```python
documents = SimpleDirectoryReader("Output").load_data()
print(f"Loaded {len(documents)} document(s).")
```

```python
embed_model = GradientEmbedding(
    gradient_access_token=os.environ["GRADIENT_ACCESS_TOKEN"],
    gradient_workspace_id=os.environ["GRADIENT_WORKSPACE_ID"],
    gradient_model_slug="bge-large",
)
```

```python
Settings.embed_model = embed_model
Settings.llm = llm
Settings.chunk_size = 1024
```

```python
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()
```

```python
response = query_engine.query(
    "What is Scaled Dot-Product Attention?"
)
print(response)
```
